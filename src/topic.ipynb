```python
import glob
import os
import pandas as pd

#Join all files

df = pd.DataFrame()

os.chdir("../data/topic")
for file in glob.glob("*.csv"):
    df_partial = pd.read_csv(file)
    df = pd.concat([df, df_partial], ignore_index=True)
df
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer_model = CountVectorizer(ngram_range=(1, 3), stop_words=stopwords_pt)

topic_model = BERTopic(embedding_model="distiluse-base-multilingual-cased-v1",
                       nr_topics=15,
                       vectorizer_model=vectorizer_model)


topic_model.fit_transform(df['text'].to_list())
```
